```python
from datlib.plots import * 
from datlib.FRED import *
data_codes  = {"1 Year": "DGS1",
              "2 Year" : "DGS2",
              "5 Year" : "DGS5",
              "10 Year" : "DGS10",
              "30 Year" : "DGS30"}
start = datetime.datetime(2001, 12, 30)
end = datetime.datetime.today()
data = gather_data(data_codes = data_codes, start = start, end = end, freq = "A")
data


def plot_lines(df, linewidth = 10, figsize = (40,20), legend = True, pp = None):
    fig, ax = plt.subplots(figsize = figsize)
    df.plot.line(linewidth = linewidth, ax = ax, legend = legend)
    ax.tick_params("both", length =10, which = "both")
    vals = ax.get_yticks()
    vals = [int(x) for x in vals]
    ax.set_yticklabels(vals)
    filename = str(list(df.keys()))
    plt.savefig(filename[:50] + "line.png",
               bbox_inches = "tight")
    if pp != None: pp.savefig(fig, box_inches = "tight")
```

```python
plt.rcParams.update({"font.size":36})
plot_lines(data)
```
